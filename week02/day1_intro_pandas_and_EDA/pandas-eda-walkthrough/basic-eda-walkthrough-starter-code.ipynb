### Description of the Boston Housing Data columns

---


This lesson uses a (famous) Boston housing market dataset to walk through a basic exploratory data analysis procedure, starting from the very beginning with loading the data. 

Though in many if not most cases the EDA procedure will be considerably more involved, this should give you an idea of the basic workflow a data scientist would go through when taking a look at a new dataset.

> **Note:** This lesson is strictly exploratory. We will not be formulating any hypotheses about the data or testing them. In many cases you may have formulated a hypothesis before even looking at your data, which could considerably affect your focus and choices in what to investigate.

The columns of the dataset have abbreviated names. The corresponding descriptions are:

    CRIM: per capita crime rate by town 
    ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
    INDUS: proportion of non-retail business acres per town 
    CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
    NOX: nitric oxides concentration (parts per 10 million) 
    RM: average number of rooms per dwelling 
    AGE: proportion of owner-occupied units built prior to 1940 
    DIS: weighted distances to five Boston employment centres 
    RAD: index of accessibility to radial highways 
    TAX: full-value property-tax rate per 10000 dollars
    PTRATIO: pupil-teacher ratio by town 
    B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town 
    LSTAT: % lower status of the population 
    MEDV: Median value of owner-occupied homes in 1000's of dollars
    
Each row in the dataset represents a different suburb of Boston.

These descriptions of shortened or coded variables are often called **codebooks** or **data dictionaries**. They are typically found with datasets you might find online in a separate file.

### 9. Standardization of variables

---

Rescaling variables is very common, and sometimes essential. Many of the machine learning algorithms that we will be using require standardized data as input.

Here we'll rescale the variables using a procedure called "standardization", which forces the distribution of each variable to have a mean of 0 and a standard deviation of 1.

Standardization is not complicated:

    standardized_variable = (variable - mean_of_variable) / std_dev_of_variable

The standardized variable is also often referred to as the **z-score** of the original variable. From the standardized variable it is easy to determine which data points are **outliers**. For example, one could say each observation with z-score larger than 3 or smaller than -3 would be more than three standard deviations away from the mean.
    
> **Note:** Nothing else is changed about the distribution of the variable. It does **not** become normally distributed.

#### Pull out the `rate of crime` and calculate its mean and standard deviation

````python
roc_mean = boston.rate_of_crime.mean()
roc_std = boston.rate_of_crime.std()
print("Mean: ", roc_mean, ", Standard deviation: ", roc_std)
````

#### Standardize the rate_of_crime variable

> Notice the new mean is centered at 0

````python
def var_standardized(var):
    var_stand = (var-var.mean())/var.std()
    return var_stand
````

````python
roc_stand = var_standardized(boston.rate_of_crime)
print("Mean: ", roc_stand.mean(), ", Standard deviation: ", roc_stand.std())
# not exactly a mean of 0 but excruciatingly close
````

#### Plot the original and standardized rate of crime

````python
fig = plt.figure(figsize=(8,4))
ax = fig.gca()

sns.distplot(boston.rate_of_crime, bins=30, kde=False)
ax.set_ylabel('counts\n', fontsize=16)
ax.set_title('Rate of crime distribution\n', fontsize=20)
plt.show()
````

````python
fig = plt.figure(figsize=(8,4))
ax = fig.gca()

sns.distplot(roc_stand, bins=30, kde=False)
ax.set_ylabel('counts\n', fontsize=16)
ax.set_title('Standardised rate of crime distribution\n', fontsize=20)
plt.show()
````

#### Logarithmic scaling

The rate of crime distribution is considerably skewed and definitely not normal. In such cases, it can often be useful to use **logarithmic scaling** for the y-axis or x-axis or both. One speaks of a semi-logarithmic or double-logarithmic plot respectively. Here we choose a logarithmic scale for the y-axis. There is no change in the variable itself. All what is changed is the visualization which makes it easier to distinguish counts which differ by many orders of magnitude.

````python
fig = plt.figure(figsize=(9,6))
ax = fig.gca()

sns.distplot(roc_stand, bins=30, hist=True, kde=False,
                  ax=ax)

ax.set_ylabel('counts\n', fontsize=16)
ax.set_title('Rate of crime distribution\n', fontsize=20)
plt.yscale('log')
plt.show()
````

We observe that apart from the first bin, the height of the bins follows roughly a straight line. An exact straight line would be observed for a variable following an exponential distribution. On a double logarithmic scale, a straight line is observed for a power-law. This is illustrated for the exponential function $\exp(-x)$ and for $x^{-2}$ below.

````python
fig, ax = plt.subplots(ncols=2, figsize=(20,6))

ax[0].plot([x for x in np.arange(0,10,0.01)],[(np.exp(-x)) for x in np.arange(0,10,0.01)])
ax[0].set_yscale('log')
ax[0].set_title('exp(-x) in semi-logarithmic scaling')

ax[1].plot([x for x in np.arange(1,1000,0.01)],[(x)**(-2.) for x in np.arange(1,1000,0.01)])
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_title('$x^{-2}$ in double-logarithmic scaling')

plt.show()
````

### 11. Covariance and correlation matrices

---

A great way to easily get a feeling for linear relationships between your variables is with a covariance or correlation matrix.

Below is the formula for the covariance between two variables $X$ and $Y$:

#### Covariance

Given sample size $N$ and variables $X$ and $Y$, with means $\bar{X}$ and $\bar{Y}$:

$$ \text{covariance}(X, Y) = \sum_{i=1}^N \frac{(X - \bar{X})(Y - \bar{Y})}{N-1}$$

The covariance is a measure of "relatedness" between variables. It is literally the sum of deviations from the mean of $X$ times deviations from the mean of $Y$ adjusted by the sample size $N$.

Code the covariance between `pct_underclass` and `home_median_value` below by hand. Verify that you got the correct result using pandas `.cov()`
on `boston[['pct_underclass','home_median_value']]`. 


> **Note:** `.cov()` returns a covariance _matrix_, which will be each value's covariance with itself and the other variable in matrix format.

````python
pct_under = boston.pct_underclass
med_val = boston.home_median_value

under_val_cov = ((pct_under - pct_under.mean())*(med_val - med_val.mean())).sum()/(len(pct_under)-1)

# Covariance between variables
print(under_val_cov)
````

````python
# Covariance matrix
np.array([[pct_under.var(),under_val_cov],
          [under_val_cov,med_val.var()]])
````

````python
boston[['pct_underclass','home_median_value']].cov()
````

````python
boston[['pct_underclass','home_median_value']].var()
````

#### Correlation

The covariance is not very interpretable. The values are difficult to read because they are relative to the variance of the variables.

A much more common metric, and one directly calculable from the covariance, is the correlation.

Again, let $X$ and $Y$ be our two variables, with covariance ${\rm cov}(X, Y)$ that we calculated above:

$$ \text{Pearson correlation}\;r = {\rm corr}(X, Y) =\frac{{\rm cov}(X, Y)}{{\rm std}(X){\rm std}(Y)}$$
