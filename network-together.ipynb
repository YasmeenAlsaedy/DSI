```python
import networkx as nx
from matplotlib import pyplot as plt

%mathplotib inline

# note: we will need more libraies 
```

```python
# undirected graphs
# create a complete graph ( every node connected )

G = nx.complete_graph(10)
nx.draw_circular(G);
```

```python
# let's make our onw graph 
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node('mynode')

# let's add some edage between the nodes
G.add_edge(1, 2)
G.add_edge(2, 'mynode')
G.add_edge(5, 6) # auto add new nodes
G.add_edge(5, 'mynode')
G.add_edge(1, 6)
```

```python
G.node(), G.edges
```

```python
nx.draw_circular(G, with_labels=True)
```

```python
print(nx.info(G))
```

```python
nx.adjacency_matrix(G).todense()
```

```python
# degree measures how many links a node has
dict(G.degree())
```

```python
# degree distribution 
plt.hist(G.degree());
```
